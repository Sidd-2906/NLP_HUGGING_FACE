```python
import torch
from auto_gptq import AutoGPTQForCausalLM
from langchain import HuggingFacePipeline, PromptTemplate
from langchain.chains import RetrievalQA
from langchain.document_loaders import PyPDFDirectoryLoader
from langchain.document_loaders import PyPDFLoader
from langchain.embeddings import HuggingFaceInstructEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import UnstructuredFileLoader
from langchain.document_loaders import DirectoryLoader
from langchain.vectorstores import Chroma
from langchain.vectorstores import FAISS
from pdf2image import convert_from_path
from transformers import AutoTokenizer, TextStreamer, pipeline
import shutil

DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
loader = PyPDFLoader('/content/indocount_aboutus.pdf')
docs = loader.load()
len(docs)
```

```python
embeddings = HuggingFaceInstructEmbeddings(
    model_name="hkunlp/instructor-large", model_kwargs={"device": DEVICE}
)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200,length_function = len)
texts = text_splitter.split_documents(docs)
len(texts)
```

```python
db = FAISS.from_documents(texts, embeddings)
```

```python
from transformers import AutoModelForCausalLM
model_name_or_path = "TheBloke/Llama-2-13B-Chat-GPTQ"
tokenizer =  AutoTokenizer.from_pretrained(model_name_or_path, use_fast=True)
model = AutoModelForCausalLM.from_pretrained(
   model_name_or_path, torch_dtype= torch.float16, trust_remote_code=True, device_map = "auto"
)
```

##Change the prompt so that unrelated data does not come in the results


```python
DEFAULT_SYSTEM_PROMPT = """
You are a Q&A assistant. Your goal is to answer questions strictly based on the provided document. If a question is outside the document's content, say "I don't know."
""".strip()

def generate_prompt(prompt: str, system_prompt: str = DEFAULT_SYSTEM_PROMPT) -> str:
    return f"""
[INST] <>
{system_prompt}
<>

{prompt} [/INST]
""".strip()
```

```python
streamer = TextStreamer(tokenizer, skip_prompt=True, skip_special_tokens=True)
```

```python
text_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=1024,
    temperature=0.0,
    top_p=0.95,
    repetition_penalty=1.15,
    streamer=streamer,
)
```

```python
llm = HuggingFacePipeline(pipeline=text_pipeline, model_kwargs={"temperature": 0.0})
```

```python
#SYSTEM_PROMPT = "Use the following pieces of context to answer the question at the end. If you don't know the answer, just say you don't know the answer or say i have no further details regarding it,please contact ameya , don't try to make up an answer."

template = generate_prompt(
    """
{context}

Question: {question}
""",
    system_prompt=DEFAULT_SYSTEM_PROMPT,
)
```

```python
prompt = PromptTemplate(template=template, input_variables=["context", "question"])
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=db.as_retriever(),
    return_source_documents=True,
    verbose=True,
    chain_type_kwargs={"prompt": prompt}


)
```

```python
lst = []
# Run the query
results = qa_chain("When was Indocount established?")

# Access the answer and source documents
answer = results['result']
source_documents = results['source_documents']

# Get the page number from the first source document (assuming it's the relevant one)
if source_documents:
    page_number = source_documents[0].metadata['page']
else:
    page_number = "N/A"  # Handle case where no source documents are available
lst.append(answer)
lst.append(page_number)
# Print the output in the desired format
print("Question: When was Indocount established?")
print(answer)
print("Source Document Page:", page_number)
```

```python
# Assuming lst is your list with the first element containing the text
text = lst[0]

# Find the index of '[/INST]'
end_index = text.find('[/INST]')

# Extract the answer text that follows '[/INST]'
if end_index != -1:
    answer_start_index = end_index + len('[/INST]')  # Start after '[/INST]'
    answer = text[answer_start_index:].strip()  # Extract and trim whitespace
else:
    answer = "I don't know."  # Handle case where '[/INST]' is not found

# Print the extracted answer
print(answer)
print("Page number is: {}".format(lst[-1]))
```

```python
import pandas as pd

# List of questions
questions = [
    "When was Indocount established?",
    "What products or services does Indocount offer?",
    "Who are the target customers of Indocount?",
    "What is the company's vision or values?",
    "How does Indocount differentiate itself from competitors?",
    "Can you provide a brief history of Indocount's growth?",
    "What are some key milestones in Indocount's development?",
    "What innovations has Indocount introduced in its field?",
    "What is Indocount's approach to quality assurance or customer service?"
]

# Initialize a list to store results
results_list = []

# Iterate over the questions
for question in questions:
    # Run the query
    results = qa_chain(question)

    # Access the answer and source documents
    answer = results['result']
    source_documents = results['source_documents']

    # Get the page number from the first source document (if available)
    page_number = source_documents[0].metadata['page'] if source_documents else "N/A"

    # Extract the answer text after '[/INST]' if it exists
    end_index = answer.find('[/INST]')
    extracted_answer = answer[end_index + len('[/INST]'):].strip() if end_index != -1 else "I don't know."

    # Append the question, answer, and page number to the results list
    results_list.append({
        'Question': question,
        'Answer': extracted_answer,
        'Source Document Page': page_number
    })

# Create a DataFrame from the results list
df = pd.DataFrame(results_list)

# Display the DataFrame
print(df)
```
